```python
import numpy as np
import cv2
import os

from scipy import ndimage
from skimage.measure import label, regionprops

%pylab inline

output_dir = os.path.join('.')
```

```python
img = cv2.imread('200x_1010_T.tif') # importar la imagen
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # pasarla a escala de grises
ret,thresh = cv2.threshold(gray,0,1,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
#ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

th1 = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,21,15) # aplicar un threshold
th2 = ndimage.gaussian_filter(th1,3) # volverlo borroso con Gauss para eliminar el ruido
th2 = cv2.adaptiveThreshold(th2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,21,5) # Volver a aplicar un threshold

#-----------------------
f, ax = plt.subplots(1, 3, figsize=(15, 3.2), sharey=True)

ax[0].imshow(img[:250,:500],cmap='gray')
ax[0].set_title('Original')

ax[1].imshow(th1[:250,:500],cmap='gray')
ax[1].set_title('Adaptive Threshold Original')

ax[2].imshow(th2[:250,:500],cmap='gray')
ax[2].set_title('Adaptive Threshold Final')

plt.tight_layout()

ax[0].axes.get_xaxis().set_visible(False)
ax[0].axes.get_yaxis().set_visible(False)
ax[1].axes.get_xaxis().set_visible(False)
ax[1].axes.get_yaxis().set_visible(False)
ax[2].axes.get_xaxis().set_visible(False)
ax[2].axes.get_yaxis().set_visible(False)

#pylab.savefig(os.path.join(output_dir,'preprocessing.png'), dpi=600)
```

```python
plt.imshow(img,cmap='gray')
plt.minorticks_on()
plt.grid()
pylab.savefig(os.path.join(output_dir,'prueba.png'), dpi=1000)
```

```python
# noise removal
kernel = np.ones((3,3),np.uint8)
close = cv2.morphologyEx(th2,cv2.MORPH_CLOSE,kernel, iterations = 2)
opening = cv2.morphologyEx(th2,cv2.MORPH_OPEN,kernel, iterations = 4)
   #opening = thresh

# sure background area
   #sure_bg = cv2.dilate(opening,kernel,iterations=1)
sure_bg = opening

# Finding sure foreground area
    #dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,0)  # dist_transform es mejor para formas definidas como monedas
    #ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
sure_fg = cv2.erode(opening,kernel,iterations=1)

# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)

#--------------------------

plt.subplot(121)
plt.imshow(opening[:250,:500],cmap='gray')
plt.title('Opening')

plt.subplot(122)
plt.imshow(thresh[:250,:500],cmap='gray')
plt.title('Original Threshold')

plt.subplots_adjust(left=0.5, bottom=0.5, right = 2.5, top=5)
```

```python
f, ax = plt.subplots(1, 2, figsize=(17, 5), sharey=True)

ax[0].imshow(sure_fg[:250,:500], cmap='gray')
ax[0].set_title('Sure Foreground - White')

ax[1].imshow(sure_bg[:250,:500], cmap='gray')
ax[1].set_title('Sure Background - Black')

plt.tight_layout()

ax[0].axes.get_xaxis().set_visible(False)
ax[0].axes.get_yaxis().set_visible(False)
ax[1].axes.get_xaxis().set_visible(False)
ax[1].axes.get_yaxis().set_visible(False)

#pylab.savefig(os.path.join(output_dir,'surefg_surebg.png'), dpi=600)
```

```python
# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)

# Add one to all labels so that sure background is not 0, but 1
markers = markers+1

# Now, mark the region of unknown with zero
markers[unknown==255] = 0
```

```python
output = img.copy()

markers = cv2.watershed(img,markers) # apply watershed
output[markers == -1] = [255,0,0] # make the borders red
```

```python
f, ax = plt.subplots(1, 2, figsize=(17, 5), sharey=True)

ax[0].imshow(img[:250,:500])
ax[0].set_title('Original')

ax[1].imshow(output[:250,:500])
ax[1].set_title('Watershed')

plt.tight_layout()

ax[0].axes.get_xaxis().set_visible(False)
ax[0].axes.get_yaxis().set_visible(False)
ax[1].axes.get_xaxis().set_visible(False)
ax[1].axes.get_yaxis().set_visible(False)

#pylab.savefig(os.path.join(output_dir,'watershed.png'), dpi=600)
```

```python
test = np.zeros((gray.shape)) # create a black background
test[markers == -1] = 1 # superpose the markers with white color

plt.imshow(test[:100,:100],cmap='gray')
#pylab.savefig(os.path.join(output_dir,'prueba2.png'), dpi=1000)
```

```python
label_img = label(test)
regions = regionprops(label_img) # Get region properties of the image
```

```python
a = []  # Create empty list

for props in regions: # Go through all the regions
    a.append(props.area) # Get the area of the regions and add it to list
    
a = np.array(a) # Turn the list into an array
a.sort() # Sort it 
a = a[np.logical_and(a>=0,a<400)]  # Remove anything smaller than or greater than certain values
```

```python
f, ax2 = plt.subplots(1, 1, figsize=(30, 10))

n, bins, patches = ax2.hist(a,bins=50,rwidth = 0.6) # Create histogram with 50 bins, and column width = 0.6 of bin width

plt.grid()
plt.yticks(np.arange(0, 425, 25))
plt.xlabel('Area [px]')
plt.ylabel('Frecuencia')

#-----------------------------------------
# Create labels for columns

# For each bar: Place a label
for rect in patches:
    # Get X and Y placement of label from rect.
    y_value = rect.get_height()
    x_value = rect.get_x() + rect.get_width() / 2

    # Number of points between bar and label. Change to your liking.
    space = 5
    # Vertical alignment for positive values
    va = 'bottom'

    # If value of bar is negative: Place label below bar
    if y_value < 0:
        # Invert space to place label below
        space *= -1
        # Vertically align label at top
        va = 'top'

    # Use Y value as label and format number with one decimal place
    label = "{:.1f}".format(x_value)

    # Create annotation
    plt.annotate(
        label,                      # Use `label` as label
        (x_value, y_value),         # Place label at end of the bar
        xytext=(0, space),          # Vertically shift label by `space`
        textcoords="offset points", # Interpret `xytext` as offset in points
        ha='center',                # Horizontally center label
        va=va)                      # Vertically align label differently for
                                    # positive and negative values.

#-------------------------------------

pylab.savefig(os.path.join(output_dir,'Hist.png'), dpi=600)
```

## Introducción

El objetivo de este taller era analizar una micrografía de un metal para intentar identificar el tamaño de grano. Para esto se aplicó el método de segmentación de imágenes conocido como Watershed Transform. Más específicamente, se utilizó una versión de ese método el cual utiliza marcadores para identificar la parte interna de lo que se está buscando detectar y así mismo identificar qué zonas no le corresponden a lo que se quiere detectar.

|![metalografia1_1.png](files/metalografia1_1.png)|![metalografia2_1.png](files/metalografia2_1.png)|
|:---------------------------------------:|:---------------------------------------:|
|**Figura 1a**. Primera metalogradía para analizar. | **Figura 1b**. Segunda metalografía para analizar.|

Este método de Watershed con marcadores internos necesita que exista un contraste fuerte entre los elementos que se van a detectar y el fondo sobre el que están. Por esto mismo se escogió sólo hacer el análisis con la Figura 1a, ya que los límites de grano están relativamente bien definidos en un color diferente al interior de los granos. Mientras que la Figura 1b tiene los granos como anillos, lo que hace que el algorítmo de Watershed se confunda y no detecte adecuadamente los granos. 

Por el tamaño de la imagen y el tamaño de los granos, de aquí en adelante se va a mostrar la imagen de la Figura 1a recortada de 0-250 y de 0-500 pixeles. 

## Pre-Processing

|![preprocessing.png](files/preprocessing.png)|
|:-------------------------------------------:|
|**Figura 2** - a: Original - b: Primer threshold - c: Segundo threshold después del Gauss Blur|

Como se mencionó anteriormente, el algorítmo necesita un contraste fuerte en la imágen para detectar los límites de los granos. Por eso el primer paso es convertir la imagen a escala de grises y aplicar un threshold para que todo quede o blanco o negro. Se aplicó un *Adaptive Gauss Threshold* (Figura 2b), después se aplicó un *Gauss Blur* para eliminar un poco de ruido, y finalmente se aplicó un segundo *Adaptive Gauss Threshold* (Figura 2c) para que las fronteras quedaran bien definidas. 

## Background & Foreground Detection

|![surefg_surebg.png](files/surefg_surebg.png)|
|:-------------------------------------------:|
|**Figura 3** - a: Sure Foreground se muestra en blanco - b: Sure Background se muestra en negro|

Con la imagen procesada se aplica un filtro morfológico para limpiarla otra vez, se identifica un área en la cúal se sabe con seguridad que es el fondo, se identifica un área en la cuál se sabe con seguridad que es el interior de los granos, y se identifica un área incierta entre estas dos. El algorítmo de Watershed lo que hará es partir del interior de los granos (la parte cierta) y empezar a crecerlos. La frontera es identifica en la parte incierta cuando dos granos creciendo choquen entre sí.

## Análisis de Granos

Ya teniendo todas las regiones divididas por el Watershed Transform, es posible utilizar una herramienta de scikit llamada *regionprops*. Lo que hace esta es identificar las regiones cerradas en una imágen y las propiedades de estas. En este caso se quiere analizar el área de los granos. 

|![Hist.png](files/Hist.png)|
|:-------------------------------------------:|
|**Figura 5**: Histograma de granos detectados. |

En este caso se limitaron las áreas entre 0px y 400px. Sin embargo, se sospecha que la manera como se calculan las áreas es contando el número total de pixeles dentro de esa área. Siendo así, ese corte entre 0 y 400 estaría incorrecto y cualquier grano por debajo de 100px podría ser considerado ruido por sobre-segmentación. 

## Conclusión 

Estas metalografías son imágenes muy complejas para un algorítmo que necesita areas muy bien definidas para poder identificarlas. De pronto con diferentes tipos de pre-procesamiento, o con otros algorítmos de detección de bordes podrían llegar a analizarse de manera detalladas imágenes como esta. Sin embargo, por límites de tiempo y conocimiento profundo de los parámetros del Watershed Transform, no fue posible profundizar más en estos algorítmos para generar un mejor análisis.
